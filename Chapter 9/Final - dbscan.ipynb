```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import DBSCAN
from collections import Counter

%matplotlib inline
plt.figure(figsize=(4, 4))

cps = pd.read_csv('cps.csv', header=None)
X = cps.iloc[:,:].values

model = DBSCAN(eps=8.0, min_samples=8).fit(X)
print(Counter(model.labels_))

cps_df = pd.DataFrame(cps)
cps_df.columns = ['FI','MU','EM', 'EV', 'S']
color_theme = cm.rainbow(np.linspace(0,1,len(Counter(model.labels_))))
plt.scatter(x=cps_df.FI,y=cps_df.EM,c=color_theme[model.labels_],s=30)
plt.xlabel(cps_df.columns[0], fontsize=12)
plt.ylabel(cps_df.columns[2], fontsize=12)
plt.show()
```
