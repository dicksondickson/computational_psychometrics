```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

%matplotlib inline
plt.figure(figsize=(4, 4))

cps = pd.read_csv('cps.csv', header=None)
X = cps.iloc[:,:].values

clustering = KMeans(n_clusters=3, random_state=8).fit(X)

cps_df = pd.DataFrame(cps)
cps_df.columns = ['FI','MU','EM', 'EV', 'S']
color_theme = np.array(['red','green','blue'])
plt.scatter(x=cps_df.FI, y=cps_df.EM, 
            c=color_theme[clustering.labels_], s=30)
plt.xlabel(cps_df.columns[0], fontsize=12)
plt.ylabel(cps_df.columns[2], fontsize=12)
plt.show()
```
